```python
words=open('words.txt','r')
words
```

```python
wordlist=words.readlines()
wordlist
```

```python
wordsclean=[word.strip().lower() for word in wordlist]
wordsclean
```

```python
wordsunique=set(wordsclean)
wordsunique
```

```python
uniqueword=list(wordsunique)
uniqueword
```

```python
uniqueword.sort()
uniqueword
```

```python
uniqword=sorted(uniqueword)
uniqword
```

```python
cleanuniqwords=sorted(list(set(word.strip().lower() for word in open('words.txt','r'))))
cleanuniqwords
```

```python
def ascword(word):
    return ''.join(sorted(word))

ascword('dabec')
```

```python
def anagrams(myword):
    return [word for word in cleanuniqwords if ascword(word)==ascword(myword)]

anagrams('act')
```

```python
%timeit anagrams('act')
```

```python
def ascwordlist():
    return {ascword(word):word for word in cleanuniqwords}

ascwordslist=ascwordlist()
ascwordslist
```

```python
def anagramsf(mywors):
    return[ascword(key) for key in ascwordslist if ascword(mywors)==ascword(key)]

anagramsf('cat')
```

```python
%timeit anagramsf('cat')
```

```python
import collections
wordsbysig=collections.defaultdict(list)
for word in cleanuniqwords:
    wordsbysig[ascword(word)].append(word)
    
wordsbysig
```

```python
def anagramsfast(mywor):
    return wordsbysig[ascword(mywor)]
    
anagramsfast('cat')
```

```python
%timeit anagramsfast('cat')
```

```python
import collections
wordsbylen=collections.defaultdict(list)
for word in cleanuniqwords:
    wordsbylen[len(word)].append(word)

wordsbylen
```

```python
swords=collections.OrderedDict(sorted(wordsbylen.items()))
swords
```

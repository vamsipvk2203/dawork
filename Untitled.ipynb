```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
countries = pd.read_csv('https://raw.githubusercontent.com/BadreeshShetty/Data-Visualization-using-Matplotlib/master/datasets/countries.csv')
```

```python
countries
```

```python
indpop=countries[countries['country']=='India']
```

```python
indpop
```

```python
indpop5=indpop.sort_values('population',ascending=True).head(10)
```

```python
indpop5
```

```python
indpop5['flag']=None
```

```python
indpop5
```

```python
for item in indpop5.index:
    if(indpop5['year'].loc[item]>1972 and indpop5['year'].loc[item]<1992):
        indpop5['flag'].loc[item]=True
```

```python
i=indpop5[indpop5['flag']==True]
i
```

```python
plt.figure(figsize=(12,6))

x=range(3)
plt.bar(x,i['population']/10**9)
plt.xticks(x,i['year'])
plt.xlabel('Countries')
plt.ylabel('Population in Billions')
plt.title('Indias population between given years 1987-2007')
plt.show()
```

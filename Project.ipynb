```python
import csv
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
listt=[]

with open('data.json') as outfile:  
    for line in outfile:
        data=json.loads(line)
        listt.append(data['data']['Author'])

listt
```

```python
import re
```

```python
li=[]
nameRegex = re.compile(r'(\w)+(\s)?[\w]+')
for lin in listt:
    author=nameRegex.search(lin)
    li.append(author.group())
li
```

```python
datauniq=set(li)
datauniq
```

```python
dict1={}
for item in datauniq:
    dict1[item]=li.count(item)
dict1
```

```python
list2=[]
list3=[]
for key,value in dict1.items():
    list2.append(key)
    list3.append(value)
list3
```

```python
df = pd.DataFrame({'name' : list2,'Commits' : list3})
```

```python
df
```

```python
df=df.sort_values('Commits',ascending=False).head(10)
```

```python
df
```

```python
plt.figure(figsize=(12,6))

x=range(10)
plt.bar(x,df['Commits'])
plt.xticks(x,df['Commits'])
plt.xlabel('Commits')
plt.ylabel('Name')
plt.title('Top 10 commits')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure(figsize=(18,6))
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/vchrombie/da-python/master/day-4/train.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.Survived.value_counts(normalize=False).plot(kind = 'bar' , alpha=0.6)
```

```python
plt.scatter(df.Survived , df.Age , alpha = 0.1)
plt.title("Age Vs Survived")
plt.show()
```

```python
df.Pclass.value_counts(normalize=True).plot(kind = "bar" , alpha = 0.6)
plt.show()
```

```python
Female_colour = "#fff000"
df.Sex[df.Survived == 1].value_counts(normalize = False).plot(kind = "bar", color = [Female_colour , 'b'])
```

```python
for x in [1]:
    df.Survived[df.Pclass == x].plot(kind = "kde")
plt.legend(("1","2","3"))
```

```python
df["hypo"] = 0
df.loc[df.Sex == "female" ,"hypo"] = 1
df
```

```python

df["result"] = 0
df.loc[df.Survived == df.hypo, "result"] = 1
df["result"].value_counts(normalize = True)
```
